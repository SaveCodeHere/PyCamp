
# Problem Set 4: Math Interpreter

- `input().split()` to parse space-separated input

- String formatting for decimal places
- Basic arithmetic operators
- Conditional statements: `if`, `elif`, `else`

---

## Problem Statement

In many programming languages, arithmetic expressions are evaluated by interpreters. Let's write a simple math interpreter that can handle basic arithmetic operations.

Write a program that:
1. Prompts the user for an arithmetic expression in the format "x y z"
   - Where x and z are numbers
   - Where y is an operator (+, -, *, /)
2. Calculates the result
3. Outputs the result formatted as a float with one decimal place

**Supported Operators:**
- `+` for addition
- `-` for subtraction
- `*` for multiplication
- `/` for division

**Input Format:**
- Three space-separated values: number operator number
- Example: "1 + 1" or "2 * 3"

**Output Format:**
- A single float number with one decimal place
- Example: 2.0 or 6.0

---

## Examples
```
Input: 1 + 1
Output: 2.0

Input: 2 * 3
Output: 6.0

Input: 6 / 2
Output: 3.0
```
